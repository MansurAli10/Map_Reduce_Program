```python
from PyPDF2 import PdfReader

# Open the PDF file
pdf_file_path = 'Harry_Potter_(www.ztcprep.com).pdf' 
pdf_reader = PdfReader(pdf_file_path)

# Extracting pages from the 7th book (birthday month is 7)
# Taking pages 8-17 (10 pages after my birth day which is the 7th)

start_page = 5957  # 8th page from book 7
end_page = 5972   # 17th page from book 7

# Initialize a string to store the extracted text (storing 8th-17th page from book 7)
extracted_text = ""

# Extract text from the specified pages
for page_number in range(start_page - 1, end_page):
    page = pdf_reader.pages[page_number]
    extracted_text += page.extract_text()

# Define the path for the output text file
output_file_path = 'C:\\Users\\Mansur\\Desktop\\603\\map reduce assign\\Doc_1.txt'

with open(output_file_path, 'w', encoding='utf-8') as output_file:
    output_file.write(extracted_text)

print(f'Extracted text saved to {output_file_path}')
```

```python
import re
from collections import defaultdict
from functools import reduce

def mapper(line):
    words = re.findall(r'\w+', line.lower())
    word_counts = defaultdict(int)
    for word in words:
        word_counts[word] += 1
    for word, count in word_counts.items():
        yield (word, count)
```

```python
def reducer(word, counts):
    total_count = sum(counts)
    return (word, total_count)

# Doc_1.txt and perform word count using map and reduce
with open('Doc_1.txt', 'r', encoding='utf-8') as file:
    lines = file.readlines()

mapped_counts = list(map(mapper, lines))

flat_mapped_counts = [item for sublist in mapped_counts for item in sublist]

word_count_dict = defaultdict(list)
for word, count in flat_mapped_counts:
    word_count_dict[word].append(count)

final_word_counts = {word: reducer(word, counts) for word, counts in word_count_dict.items()}

for word, count in final_word_counts.items():
    print(f'{word}: {count}')
```

```python
import re
from collections import defaultdict
from functools import reduce

# Load the list of English words from the file extracted from (http://www.gwicks.net/dictionaries.htm) 
english_words_file = 'english_predefined_words.txt'
with open(english_words_file, 'r', encoding='utf-8') as file:
    english_words = set(word.strip().lower() for word in file.readlines())

def is_english(word):
    return word.lower() in english_words

# Mapper function to remove unwanted characters and count non-English words
def map_non_english_words(line):
    # Use regular expressions to remove unwanted characters adjacent to words
    words = re.findall(r'\b\w+\b', line.lower())
    non_english_words = {}
    for word in words:
        if not is_english(word):
            non_english_words[word] = non_english_words.get(word, 0) + 1
    return non_english_words
```

```python
# Reducer function to combine non-English word counts
def reduce_non_english_words(count1, count2):
    for word, count in count2.items():
        count1[word] = count1.get(word, 0) + count
    return count1

with open('Doc_2.txt', 'r', encoding='utf-8') as file:
    lines = file.readlines()

mapped_non_english_words = list(map(map_non_english_words, lines))

final_non_english_words = defaultdict(int)
for word_count in mapped_non_english_words:
    final_non_english_words = reduce(reduce_non_english_words, [final_non_english_words, word_count])

for word, count in final_non_english_words.items():
    print(f'{word}: {count}')
```
